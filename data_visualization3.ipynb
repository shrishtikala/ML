```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```
#reading data fromm csv
df=pd.read_csv('http://13.234.66.67/summer19/datasets/bank.csv')
```

```
df
```

```
df.info()
```

```
top=df.head(5)
```

```
top
```

```
#ploting piechart
cs=top['CreditScore']
s=top['Surname']
exp=(0,0,0,0,0.1)

plt.pie(cs,labels=s,shadow=True,explode=exp,autopct='%1.1f%%')
```

```
plt.bar(df.Surname,df.EstimatedSalary)
plt.xlabel('Surname')
plt.ylabel('Salary')
```

```
plt.scatter(df.Age,df.EstimatedSalary)
plt.xlabel('Age')
plt.ylabel('Salary')
```

```
plt.xlabel('Gender')
plt.ylabel('Age')
plt.bar(df.Gender,df.Age)
```

```
df.groupby(['Geography']).size().plot('pie',shadow=True,label='Geography',autopct='%1.1f%%')
```

```
df.groupby(['Gender']).size().plot('pie',shadow=True,label='Gender',autopct='%1.1f%%')
```
